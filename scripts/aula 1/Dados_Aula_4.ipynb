```python
# Indexação
## ìndices de listas
compras = ['leite','comida molhada','ração']
compras[1]

## indexação de dicionários

moradores = {'um':{'nome':'Xerxes','idade':16},'dois':{'nome':'Calvin','idade':6}}
moradores['um']['idade']
```

```python
# Indexação de DataFrames
# Para acessar as colunas, faça como se o título da coluna fosse uma chave em um dicionário
dados['Título']
# Para acessar os elementos, indexe como lista
dados['Título'][0:10]
```

```python
# Operações de comparação e identidade em DataFrmes
## Obter os índices em que a condição é verdadeira
dados['Título'] == 'Herói de brinquedo'
```

```python
# Selecionar elementos a partir de uma condição
dados[dados['Título'] == 'Herói de brinquedo']
```

```python
dados['Que nota daria para o filme?'] == 5
```

```python
# Selecionar elementos a partir de um teste de comparação
dados[dados['Que nota daria para o filme?'] == 5]
```

```python
# Quantas observações? 
len(dados)
dados.shape
```

```python
# Quantos filmes receberam cada nota?
## A função groupby()

dados.groupby('Que nota daria para o filme?').count()
```

```python
dados.groupby('Categoria').count().sort_values('Carimbo de data/hora',ascending=False)
```

```python
# Ordenação de valores
## Funções sorted e sort_values

dados_ordenado = sorted(dados['Que nota daria para o filme?'])
```

```python
dados_ordenado
```

```python
## sort_values

dados.sort_values('Título')
```

```python
# Estatística descritiva
## Tendências centrais
## Média
dados.mean()
```

```python
## Mediana 
dados.median()
```

```python
# Dispersão
## Amplitude
dados['Que nota daria para o filme?'].max() - dados['Que nota daria para o filme?'].min()
```

```python
## Variância e desvio padrão

dados['Que nota daria para o filme?'].var()
```

```python
dados['Que nota daria para o filme?'].std()
```

```python
dados['Comprimento Título'] = [len(i) for i in dados['Título']]
dados
```

```python
# Correlação
## Covariância

dados[['Que nota daria para o filme?','Comprimento Título']].sort_values('Comprimento Título',ascending=False)
dados[['Que nota daria para o filme?','Comprimento Título']].cov()
```

```python
dados[['Que nota daria para o filme?','Comprimento Título']].corr()
```

```python
# A função describe gera um sumário de estatística descritiva

dados.describe()
```
